## Plot MLP's confusion matrix

```python
from mlxtend.plotting import plot_confusion_matrix
import matplotlib.pyplot as plt
import numpy as np

binary = np.array([[8070,1180],
                   [1050, 8200]])

fig, ax = plot_confusion_matrix(conf_mat=binary,
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True)
plt.show()
#fig.savefig("mlpmatrix_lemma.pdf")
```

## Plot the logistic sigmoid function

```python
import math

def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a
```

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)
```

```python
f = plt.figure()
plt.plot(x,sig)
plt.show()

#f.savefig("sigmoid.pdf")
```
